# Cost Function vs. Loss Function

Often used interchangeably, but subtle differences exist:

- **Loss function:** Measures error per individual sample (data point).
- **Cost function:** Aggregates loss over the entire dataset.

### A. Regression
- **Mean Squared Error (MSE):** Simple and interpretable, but sensitive to outliers.
- **Mean Absolute Error (MAE):** Robust to outliers, but less intuitive interpretation.
- **Huber Loss:** Combines benefits of MSE and MAE, less sensitive to outliers than MSE.

# Regression Loss

## 1. Mean Squared Error/Squared loss/ L2 loss

The Mean Squared Error (MSE) is the simplest and most common loss function. To calculate the MSE, you take the difference between the actual value and model prediction, square it, and average it across the whole dataset.



image.png

**Advantages:**

1. Easy to interpret.
2. Always differentiable because of the square.
3. Only one local minima.

**Disadvantages:**

1. Error unit in the square, because the unit in the square is not understood properly.
2. Not robust to outliers.

**Note:** In regression, at the last neuron use linear activation function.


## 2. Mean Absolute Error/ L1 loss

The Mean Absolute Error (MAE) is also a simple loss function. To calculate the MAE, you take the difference between the actual value and model prediction and average it across the whole dataset.





image.png

**Advantages:**

1. Intuitive and easy.
2. Error unit same as the output column.
3. Robust to outliers.

**Disadvantages:**

1. The graph is not differentiable. We cannot use gradient descent directly, then we can use subgradient calculation.

**Note:** In regression, at the last neuron use linear activation function.

### B. Classification
- **Binary Cross-Entropy:** Used for binary classification (two classes), penalizes predictions based on their distance from the true value.
- **Categorical Cross-Entropy:** Used for multiclass classification, measures the cross-entropy between predicted probabilities and true labels.

# Cross-Entropy

Cross-entropy loss, often simply called "cross-entropy," is a crucial loss function used in machine learning, especially in classification problems.


It measures the dissimilarity between two probability distributions - the actual distribution of labels in the data and the predicted distribution of labels by the model.

.


# B. Classification Loss

## 1. Binary Cross Entropy/Log Loss

It is used in binary classification problems, such as determining whether a person has COVID or not, or whether an article becomes popular or not.

Binary cross entropy compares each of the predicted probabilities to the actual class output, which can be either 0 or 1. It calculates a score that penalizes the probabilities based on the distance from the expected value, indicating how close or far they are from the actual value.

## 2. Categorical Cross Entropy

Categorical Cross Entropy is used for multiclass classification and softmax regression.

**Loss function:** `-sum up to k(y_j log(_j))` where k is the number of classes.



image.png

**Cost function:** `-1/n * sum up to n(sum from j to k (y_ij log(_ij)))`

where:
- **k** is the number of classes,
- **y** = actual value,
- ****  Neural Network prediction.



image.png

**Note:** In multi-class classification, at the last neuron use the softmax activation function.

**Softmax Activation:** For a problem statement with 3 classes, the softmax activation function is `f(z) = e^(z1) / (e^(z1) + e^(z2) + e^(z3))`.

### When to use Categorical Cross-Entropy and Sparse Categorical Cross-Entropy?

- Use **Categorical Cross-Entropy** if the target column has One-hot encoding of classes like `0 0 1`, `0 1 0`, `1 0 0`.
- Use **Sparse Categorical Cross-Entropy** if the target column has numerical encoding of classes like `1, 2, 3, 4n`.

### Which is Faster?

Sparse Categorical Cross-Entropy is faster than Categorical Cross-Entropy.
