# Forward Propagation in Neural Networks

Forward propagation in neural networks is the process by which input data is transformed into an output through successive layers of the network. This document provides an example to illustrate this concept.

## Conceptual Overview

- **Input Data**: Begins at the input layer.
- **Transformation**: Each neuron in the subsequent layers processes the data.
- **Output**: The final layer produces the network's output.

## Mathematical Explanation

- **Neuron Calculation**: Each neuron calculates a weighted sum of its inputs, adds a bias, and then applies an activation function.
  - **Formula**: `output = activation(weight * input + bias)`.
- **Layer-wise Computation**: This process is repeated layer by layer until the output layer is reached.

## Example

Consider a simple neural network with:
- 2 input neurons (features)
- 1 hidden layer with 2 neurons
- 1 output neuron

The activation function used is the Rectified Linear Unit (ReLU), defined as `ReLU(x) = max(0, x)`.

### How Forward Propagation Works for a Single Data Point

1. **Input Layer**:
   - Receives the input `[x1, x2]`.

2. **Hidden Layer**:
   - Neuron 1 computes `output1 = ReLU(w11 * x1 + w12 * x2 + b1)`.
   - Neuron 2 computes `output2 = ReLU(w21 * x1 + w22 * x2 + b2)`.

3. **Output Layer**:
   - Computes final output `output = ReLU(w31 * output1 + w32 * output2 + b3)`.

Where `w` are weights, and `b` are biases for each neuron.




Let's implement this in Python:

````python
import numpy as np
````

````python
input_data = np.array([1,2])
w1 = np.array([[0.1,0.2],[0.3,0.4]])
b1 = np.array([0.1,0.2])
w2 = np.array([0.1,0.2])
b2 = np.array(0.1)

# Activation Function
def relu(x):
    return np.maximum(0, x)

````

````python
# Hidden layer
hidden_layer_input = np.dot(w1,input_data) +b1
hidden_layer_output = relu(hidden_layer_input)
````

````python
#output Layer
output_layer_input = np.dot(w2,hidden_layer_output) + b2
output = relu(output_layer_input)
print(output)
````

````python
# import required library
import numpy as np

# Sample input
input_data = np.array([1,2])

# Sample weights and biases
w1 = np.array([[0.1, 0.2],[0.3, 0.4]]) # weight for hidden layer
b1 = np.array([0.1, 0.2])          # biases for hidden layer
w2 = np.array([0.1, 0.2])            # weight for output layer
b2 = np.array(0.1)                 # biases for output layer

# Activation Function / threshold
def relu(x):
    return np.maximum(0, x)

# Forward Propagation
# Hidden Layer
hidden_layer_input = np.dot(w1, input_data)+b1
hidden_layer_output = relu(hidden_layer_input)

# Output Layer
output_layer_input = np.dot(w2 , hidden_layer_output) + b2
output = relu(output_layer_input)

print(output)
````
